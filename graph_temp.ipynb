### 주사위 시뮬레이션
* 임의의 수를 뽑는 랜덤 함수를 활용하여 주사위 시물레이션 만들기
* 주사위를 굴림 -> 결과를 기록 -> 앞의 과정을 n번 반복 -> n+1번의 결과를 히스토그램으로 그리기

```python
# in range(5) :5번 반복한다는 의미
import random
dice=[]
for i in range(5):
    dice.append(random.randint(1, 6))
print(dice)
```

```python
# dice리스트를 히스토그램으로 표현하기
# bins : x축의 개수 설정
import random
dice=[]
for i in range(5):
    dice.append(random.randint(1, 6))
print(dice)

plt.hist(dice, bins=6)
plt.show()
```

```python
# dice리스트를 히스토그램으로 표현하기
# bins : x축의 개수 설정
import random
dice=[]
for i in range(100):
    dice.append(random.randint(1, 6))
print(dice)

plt.hist(dice, bins=6)
plt.show()
```

```python
# dice리스트를 히스토그램으로 표현하기
# bins : x축의 개수 설정
import random
dice=[]
for i in range(1000000):
    dice.append(random.randint(1, 6))

plt.hist(dice, bins=6)
plt.show()
```

### 기온데이터 시각화 하기

```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1] !='':
        result.append(float(row[-1]))

plt.hist(result, bins=100, color='g')
plt.show()
```

```python
# 8월 데이터만 뽑아서 히스토 그램 확인하기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
         if month=='08':
                aug.append(float(row[-1]))
                
plt.hist(aug, bins=100, color='b')
plt.show()
```

```python
# 그래프 타이틀에 한글 폰트 오류 수정
plt.rc('font', family="Malgun Gothic")
# 그래프에 음수 오류 수정
plt.rcParams["axes.unicode_minus"]=False
```

```python
# 1월과 8월 동시에 히스토 그램 그리기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1] != '':
        if month=='01':
            jan.append(float(row[-1]))
        if month=='08':
            aug.append(float(row[-1]))

plt.hist(jan, bins=100, color='g', label="1월")
plt.hist(aug, bins=100, color='orange', label="8월")
plt.legend()
plt.show()
```

```python
# 서울 최고기온 데이터를 boxplot으로 그리기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data :
    if row[-1] != '':
        result.append(float(row[-1]))
        
plt.boxplot(result)
plt.show()
```

```python
# 1월과 8월의 최고 기온 데이터를 boxplot 으로 그리기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == "01":
            jan.append(float(row[-1]))
        if month == "08":
            aug.append(float(row[-1]))
plt.boxplot([jan, aug])
plt.show()        
# 표에 나타나는 동그라미는 이상치
```

```python
# 월별 최고기온을 boxplot으로 그리기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
month=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != "":
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
# 5~ 9월까지는 편차가 작고, 1~4월, 10~12월은 편차가 큰 것 같음. 특히 12월이 편차가 큰 것으로 보임
```

```python
# 8월 일별 기온 데이터를 boxplot으로 그리기
f=open('seoul.csv')
data=csv.reader(f)
next(data)

day=[]
# day 리스트 안에 31개의 리스트 생성
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=="08":
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

# 그래픽 스타일 지정
plt.style.use('ggplot')
# 그래픽 크기 지정
plt.figure(figsize=(10, 5), dpi=300)
# showfilters=False : 이상치 제거
plt.boxplot(day, showfliers=False)

plt.show()
```
