# Numpy 활용하기
* 파이썬 리스트를 이용하는 것보다 적은 수의 코드로 작성할 수 있음

```python
import matplotlib.pyplot as plt
t=[]
p2=[]
p3=[]

for i in range(0, 50, 2) :
    t.append(i /10)
    p2.append((i /10)**2)
    p3.append((i /10)**3)

plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
import numpy as np
t =np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## numpy 익히기

```python
import numpy as np
print(np.sqrt(2))
print(np.pi)
print(np.sin(0))
print(np.cos(np.pi))
```

```python
# rand() : 0과 ` 사이 중에서 n개의 실수가 랜덤으로 생성
# nd : N-Dimensional, array : 배열
import numpy as np
a=np.random.rand(5)
print(a)
print(type(a))
```

```python
#  0~5 사이 숫자를 랜덤하게 10번 선택
print(np.random.choice(6, 10))
# replace=False : 한 번 뽑은 숫자를 다시 뽑지 못함
# 0~9 사이 숫자를 랜덤으로 중복없이 6번 뽑음
print(np.random.choice(10, 6, replace=False))
# p속성을 통해 확률 지정 가능
print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))
```

### numpy를 활용해서 그래프 그리기

```python
# graph_temp 에서 히스토그램 시각화를 위해 사용한 코드
import random
dice=[]
for i in range(5):
    dice.append(random.randint(1, 6))
print(dice)

plt.hist(dice, bins=6)
plt.show()
```

```python
# graph_temp 에서 그렸던 코드와 동일한 결과 but 더 간결한 코드
# 0~5 사이의 숫자를 랜덤으로 10개 선정한 후, 그 데이터를 이용해서 가로축이 6개인 히스토그램 그래프 그리기
dice=np.random.choice(6, 10)

plt.hist(dice, bins=6)
plt.show()
```

```python
# 반복 횟수 100만번 및 확률 지정
dice=np.random.choice(6, 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])

plt.hist(dice, bins=6)
plt.show()
```

```python
# people 에서 버블차트 시각화한 코드
import random
x=[]
y=[]
size=[]

for i in range(100):
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.5)
plt.colorbar()
plt.show()
```

```python
#  people 버블 차트와 동일한 결과
# 코드가 훨씬 간결함
x=np.random.randint(10, 100, 100)
y=np.random.randint(10, 100, 100)
size=np.random.rand(100)*100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.5)
plt.colorbar()
plt.show()
```

### numpy array 생성하기
* ndarray를 다룰 것

```python
# , 는 ndarray 리스트 내 요소를 구분하는 기호
# 출력값을 보면 []는 출력되지만 ,는 출력되지 않음

a=np.array([1, 2, 3, 4])
print(a)
```

```python
a=np.array([1, 2, 3, 4])
print(a[1], a[-1]) # 특정 값만 출력 가능
print(a[1:]) # 특정 값 기준으로 슬라이싱하여 출력
```

```python
# numpy는 한 가지 타입의 데이터만 저장 가능. 숫자+문자로 이루어진 데이터는 숫자를 문자로 변환하여 저장함
a=np.array([1, 2, '3', 4])
print(a)
```

```python
# zeros(), ones(), eye() 등을 이용해서 배열울 초기화할 수 있음
a=np.zeros(10) # 0으로 이루어진 크기가 10인 배열 생성
a1=np.ones(10) # 1로 이루어진 크기가 10인 배열 생성
a2=np.eye(3) # 3*3 단위의 배열 생성
print(a)
print(a1)
print(a2)
```

```python
# arrange()를 통해 0과 1이 아닌 연속된 숫자 생성도 가능
print(np.arange(3)) # arange() 하나의 값만 입력할 경우, 0보다 크면서 해당 숫자보다 1 작은 수까지의 배열 생성
print(np.arange(3, 7)) # arange() 두개의 값을 입력할 경우, 첫번째 값보다 크면서 두번째 값보다 1 작은 수까지의 배열 생성
# arange() 세개의 값을 입력할 경우,
#첫번째 값보다 크면서 두번째 값보다 1 작은 수의 범위 내에서, 세번째 숫자 만큼의 간격을 둔 배열 생성
print(np.arange(3, 7, 2))
```

```python
# 또 다른 숫자 생성 함수 linspace(), 이 함수의 경우 구간의 갯수를 지정할 수 있음
# 특성에 맞게 사용할 것
a=np.arange(1, 2, 0.1)
b=np.linspace(1, 2, 11) # 1부터 2까지 11개의 구간으로 나눈 실수 생성

print(a)
print(b)
```

```python
a=np.arange(-np.pi, np.pi, np.pi/10)
b=np.linspace(-np.pi, np.pi, 20)

print(a)
print(b)
```
