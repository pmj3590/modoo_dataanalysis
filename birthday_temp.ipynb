# 기본 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40])
plt.title("plotting")
plt.show()
```

```python
# plt.plot() : 안에 리스트가 2개 일 때, 전자의 경우는 x축, 후자는 y축이됨.
# 이 때, x축과 y축에 해당하는 개수는 동일해야 함.
plt.plot([1, 2, 3,4], [12,43, 25,15])
plt.show()
```

```python
plt.title("plotting")
# label ; 범례의 이름 지정
plt.plot([10,20,30,40], label="asc")
plt.plot([40,30,20,10], label="desc")
# plt.legend() : 표에 범례 추가
plt.legend()
plt.show()
```

```python
plt.title("color")
plt.plot([10,20,30,40], color="green", label="green")
plt.plot([40,30,20,10], color="pink", label="pink")
plt.legend()
plt.show()
```

```python
plt.title('linestyle')
plt.plot([10,20,30,40], color="red", linestyle="--", label="dashed")
plt.plot([40,30,20,10], color="blue", linestyle=":", label="dotted")
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label="circle")
plt.plot([40,30,20,10], 'g^', label="triangle up")
plt.legend()
plt.show()
```

# 내 생일의 기온 변화를 그래프로 그리기

```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
for row in data:
    if row[-1] !='':
        result.append(float(row[-1]))
    
print(result)
```

## 최고 기온 중 날짜 데이터만 추출하기
* split을 이용하여 날짜 데아터 추출하기

```python
date='2015-04-21'
print(date.split('-'))
print(date.split('-')[0])
print(date.split('-')[1])
print(date.split('-')[2])
```

```python
# 8월의 최고 기온 데이터만 추출하기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
for row in data:
    if row[-1] !='':
        if row[0].split('-')[1]=='08':
            result.append(float(row[-1]))
plt.plot(result, 'green')
plt.show()
```

```python
# 매년 2월 14일 데이터 추출하기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
for row in data:
    if row[-1] !='':
        if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='14':
            result.append(float(row[-1]))
plt.plot(result, 'green')
plt.show()
```

```python
# 그래프 타이틀에 한글 폰트 오류 수정
plt.rc('font', family="Malgun Gothic")
# 그래프에 음수 오류 수정
plt.rcParams["axes.unicode_minus"]=False
```

```python
# 내 생일인 날짜의 최저, 최고 기온 데이터 추출하기
f=open('seoul.csv')
data=csv.reader(f)
next(data)
# 최고기온 값을 저장할 리스트 생성
high=[]
# 최저기온 값을 저장할 리스트 생성
low=[]
for row in data:
    if row[-1] !='' and row[-2] !='':
        if 2015 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1]=='02' and row[0].split('-')[2]=='21':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.plot(high, 'violet', label="최고 기온")
plt.plot(low, 'yellow', label="최저 기온")
plt.title('내 생일의 기온 변화')
plt.legend()
plt.show()
```
